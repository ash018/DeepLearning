#### In this notebook we will cover the basics of machine learning. Here we will see the concept of generalizing over unseen example or the test sets, overfitting & underfitting of training data, the capacity of the model & the notion of regularization. 

```python
#Generate Dataset
%matplotlib inline
import pylab
import numpy
```

```python
x = numpy.linspace(-1,1,100) # Generating equidistant 100 data between -1 to +1
signal = 2 + x + 2*x*x # output function
noise = numpy.random.normal(0,0.1,100) # adding noise
y = signal + noise # finding output variable
```

```python
pylab.plot(signal,'b')
pylab.plot(y,'g')
pylab.plot(noise,'r')
pylab.xlabel("x")
pylab.ylabel("y")
pylab.legend(["Without Noise","With Noise","Noise"],loc=2)

#Training set
x_train = x[0:80]
y_train = y[0:80]

pylab.figure()
degree = 2
X_train = numpy.column_stack([numpy.power(x_train,i) for i in range(0,degree)])
model = numpy.dot(numpy.dot(numpy.linalg.inv(
    numpy.dot(X_train.transpose(),X_train)),X_train.transpose()),y_train)

pylab.plot(x,y,'g')
pylab.xlabel("x")
pylab.ylabel("y")

predicted = numpy.dot(model,[numpy.power(x,i) for i in range(0,degree)])
pylab.plot(x,predicted,'r')
pylab.legend(["Actual","Predicted"],loc=2)
```

```python
#training & testing Root MEAN square error value
train_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[0:80] - predicted[0:80],y_train - predicted[0:80])))
test_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[80:] - predicted[80:], y[80:] - predicted[80:])))
print("Train RMSE (Degree = 1)", train_rmse1)
print("Test RMSE (Degree = 1)", test_rmse1)
```

```python
# Model with degree 2

pylab.figure()
degree = 3
X_train = numpy.column_stack([numpy.power(x_train,i) for i in range(0,degree)])
model = numpy.dot(numpy.dot(numpy.linalg.inv(
    numpy.dot(X_train.transpose(),X_train)),X_train.transpose()),y_train)

pylab.plot(x,y,'g')
pylab.xlabel("x")
pylab.ylabel("y")

predicted = numpy.dot(model,[numpy.power(x,i) for i in range(0,degree)])
pylab.plot(x,predicted,'r')
pylab.legend(["Actual","Predicted"],loc=2)

#training & testing Root MEAN square error value
train_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[0:80] - predicted[0:80],y_train - predicted[0:80])))
test_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[80:] - predicted[80:], y[80:] - predicted[80:])))
print("Train RMSE (Degree = 1)", train_rmse1)
print("Test RMSE (Degree = 1)", test_rmse1)
```

```python
# Model with degree 8

pylab.figure()
degree = 9
X_train = numpy.column_stack([numpy.power(x_train,i) for i in range(0,degree)])
model = numpy.dot(numpy.dot(numpy.linalg.inv(
    numpy.dot(X_train.transpose(),X_train)),X_train.transpose()),y_train)

pylab.plot(x,y,'g')
pylab.xlabel("x")
pylab.ylabel("y")

predicted = numpy.dot(model,[numpy.power(x,i) for i in range(0,degree)])
pylab.plot(x,predicted,'r')
pylab.legend(["Actual","Predicted"],loc=2)

#training & testing Root MEAN square error value
train_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[0:80] - predicted[0:80],y_train - predicted[0:80])))
test_rmse1 = numpy.sqrt(numpy.sum(numpy.dot(y[80:] - predicted[80:], y[80:] - predicted[80:])))
print("Train RMSE (Degree = 1)", train_rmse1)
print("Test RMSE (Degree = 1)", test_rmse1)
```

#### Let's discuss the concept of model capacity. The data were generated using second order polynomial (degree = 2) with same noise. Then we tried to approximate the data using three models of degree: 1,2 and 8 respectively. The higher the degree the model can accommodate more variation. This ability to accommodate variation corresponds to the notion of capacity. Thus we can say the model with degree = 8 has higher capacity than the model with degree = 2 which in turn has higher capacity than degree = 1. But is having higher capacity is good? It turns out not.

#### We can see the model with degree = 2 performs better in unseen data than degree = 8 . We can say that model with degree = 1 suffers from underfitting and degree = 8 suffers from overfitting. 

#### A regularized version of least squares takes the form  y=$\beta$x where $\beta$ is a vector such that  ||X$\beta$-y||$^2$ + $\lambda$ ||$\beta$||$^2$ is minimized and $\lambda$ is a user defined parameter that controls the complexity.

```python
#Regularization


x_train = x[0:80]
y_train = y[0:80]

train_rmse = []
test_rmse = []
degree = 80
lambda_reg_values = numpy.linspace(0.01,0.99,100)

for lambda_reg in lambda_reg_values:
    X_train = numpy.column_stack([numpy.power(x_train,i) for i in range(0,degree)])
    model = numpy.dot(numpy.dot
                      (numpy.linalg.inv
                                (numpy.dot(X_train.transpose(),X_train) + 
                                                 lambda_reg * numpy.identity(degree)),X_train.transpose()),y_train)
    
    predicted = numpy.dot(model,[numpy.power(x,i) for i in range(0,degree)])
    
    train_rmse.append(numpy.sqrt(numpy.sum(numpy.dot(y[0:80] - predicted[0:80],
                                                   y_train - predicted[0:80] ))))
    test_rmse.append(numpy.sqrt(numpy.sum(numpy.dot(y[80:] - predicted[80:],
                                                   y[80:] - predicted[80:] ))))
    
pylab.plot(lambda_reg_values,train_rmse)
pylab.plot(lambda_reg_values,test_rmse)
pylab.xlabel(r"$\lambda$")
pylab.ylabel("RMSE")
pylab.legend(["Train","Test"],loc=2)
```

#### So here lambda value less 0.35 occurred under-fitting & greater than that caused over-fitting
